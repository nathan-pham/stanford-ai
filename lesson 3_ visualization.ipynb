# Random walk
The story of drunkard walker: There once was a drunk man who wandered in the middle of nowhere. He randomly picked a direction towards east,west, south or north. We would like to see where he's going to end up after a few steps. 

```python
from random import randint

def random_walk(nstep: int):
    """
    This function carry out a random walk which contains nstep random walk
    Every step, the walker moves randomly in North, South, East or West direction.
    The random walk starts at (0,0)
    inputs:
    nstep: number of steps in the random walk. 

    output:
    x_list, y_list: lists of x and y coordinates of the walker.
    """

    points = {
        0: [0, 1],
        1: [0, -1],
        2: [1, 0],
        3: [-1, 0]
    }

    x, y = [0], [0]

    for i in range(nstep - 1):
        x_coord, y_coord = points.get(randint(0, 3))

        x.append(x[i] + x_coord)
        y.append(y[i] + y_coord)

    return x, y

# random_walk(6)
```

Now run and visualize random walk


```python
nstep = 500
x_list, y_list = random_walk(nstep)

# plot x_list,y_list using plt.scatter
plt.figure(figsize=(8, 8))
plt.scatter(x_list, y_list, marker='o', s=50, c=range(nstep));
```
